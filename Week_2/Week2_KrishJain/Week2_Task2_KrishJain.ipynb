```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Drop Setosa
df.drop(df[df['Species']=='Iris-setosa'].index, inplace=True)

# Assign value to versi and viriginica
df.loc[df["Species"] == "Iris-versicolor", "Species"] = 0
df.loc[df["Species"] == "Iris-virginica", "Species"] = 1

# Lets check the df again
display(df)
```

```python
# Forming training and test set, lets keep test set to be 20 percent
Y = df.Species
aspects = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
X = df[aspects]
train_X, test_X, train_Y, test_Y = train_test_split(X,Y,test_size = 0.20,random_state = 1)
```

```python
# Initialize the model
learning_rate = 0.001
epochs = 10000
slopes = np.zeros(4,dtype='float64')
intercepts = 0
entries, n_aspects = train_X.shape
```

```python
# Regression

def sigmoid(arr):
  for i in range(len(arr)):
    arr[i] = 1/(1+np.exp(-float(arr[i])))
  return arr

def train(X, Y, learning_rate, epochs, slopes, intercepts):
  global entries

  for i in range(epochs):
      # Predicted values
      predicted_Y = np.dot(X,slopes) + intercepts
      predicted_final = sigmoid(predicted_Y)

      # Gradient Descend
      d_slopes = (1/entries)*np.dot(X.T,predicted_final-Y)
      d_intercepts = (1/entries)*np.sum(predicted_final - Y)


      slopes = slopes - learning_rate*d_slopes
      intercepts = intercepts - learning_rate*d_intercepts

  return slopes, intercepts

def test(X, Y, slopes, intercepts):
    predicted_Y = np.dot(X, slopes) + intercepts
    predicted_final = sigmoid(predicted_Y)

    return predicted_final

def cap(Y):
  for i in range(len(Y)):
      if Y[i] < 0.5:
        Y[i] = 0
      else:
        Y[i] = 1

  return Y

def accuracy(Y,predicted_Y):
  return (np.sum(Y==predicted_Y))*100/len(Y)

```

```python
slopes , intercepts = train(train_X, train_Y, learning_rate, epochs, slopes, intercepts)
response = test(test_X, test_Y, slopes, intercepts)
response = cap(response)
```

```python
# Checking accuracy
print(accuracy(test_Y,response))
```
