```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
w=np.zeros((X_train.shape[1],1))
```

```python
def logistic_reg(X,Y,epoch,rate,n,w):
  #w=np.zeros((X.shape[1],1))
  #b=np.zeros((X_t.shape[0],1))
  for i in range(epoch):
    z=np.dot(X,w)
    y_pred=sigmoid(z)
    dw =(-1/n) * np.dot(X.T,(Y-y_pred))
    #db=(1/n) * np.sum(Y-y_pred)
    w=(w)-(rate)*(dw)
    #b=b-rate*db
  return w
```

```python
def prediction(X_t,w):
   z=np.dot(X_t,w)
   y_pred=sigmoid(z)
   ans=[0 if y<=0.5 else 1 for y in y_pred]
   return ans
```

```python
#np.zeros((X.shape[1],1))
weights=logistic_reg(X_train,Y_train,100000,0.001,80,w)
ans=prediction(X_test,weights)
print(ans)
```

```python
def accuracy (y_pred, y_test):
  return np. sum (y_pred==y_test) /len (y_test)
```

```python
import sklearn
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, ans)
print(accuracy)
```
