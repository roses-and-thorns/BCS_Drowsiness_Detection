```python
# Importing Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate Random Data
n = 1000
X = 10* np.random.random(n) - 5
Y = 4*X + 3 + np.random.normal(loc =0, scale = 10, size = n)

plt.figure()
plt.scatter(
    X,
    Y
)
```

```python
def gradient_descent(a,b,rate,iter):
  for i in range(iter):
    z=a*X+b
    mse=(1/n)*np.sum((Y-z)**2)
    grad=(-2/n)*np.dot(X.T,(Y-z))
    grad1=(-2/n)*np.sum(Y-z)
    a=a-rate*grad
    b=b-rate*grad1


  return a,b

```

```python
def Y_based_on_X(x,a,b):
  return a*x +b

# Gradient Descent
l=0.001
a,b=gradient_descent(0,0,l,1000)
print(a,b)


plt.figure()
plt.scatter(
    X,
    Y
)

plt.scatter(X,Y_based_on_X(X,a,b),c = "red")
```
